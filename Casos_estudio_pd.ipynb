# Caso practico 1

```python
# Importar datos de trabajo
import pandas as pd
df_raw = pd.read_csv('bikes2018.csv',
                     sep = ',',
                     encoding = 'utf8',
                     parse_dates = ['Date'],
                     dayfirst = True,
                     index_col = 'Date')
df_raw.head()
```

```python
# Obtener algunas estadisticas descriptivas
df_raw.describe()
```

```python
# Seleccionar solo los flujos de bicicletas de los carriles de interes
df_selected = df_raw[['Berri1','Boyer','Boyer 2', 'Parc', 'PierDup', 'Viger']]
df_selected
```

```python
# Manejo de los valores NaN
df_selected.fillna(value = 0, inplace = True)
```

```python
# Conteo del flujo total de autos por calle en el periodo analizado
df_selected.sum().sort_values(ascending = False)
```

```python
# Detectar calle con mayor flujo anual
print('La calle con mayor flujo es {} con {} bicicletas.'.format(
    df_selected.columns[df_selected.sum().argmax()]  ,  
    df_selected[df_selected.columns[df_selected.sum().argmax()]].sum()
))
```

```python
# Grafico de barras
df_selected.sum().plot(kind = 'bar')
```

```python
# Mes en el que mas ciclistas circulan

# Agregar fechas como columna para analisis
df_selected.reset_index(inplace = True)
```

```python
# Creacion de un dataframe copia
df = df_selected.copy()
df['Year'] = df['Date'].dt.year 
df['Month'] = df['Date'].dt.month
df['Total'] = df.Berri1 + df.Boyer + df['Boyer 2'] + df.Parc + df.PierDup

df.head(5)
```

```python
df.groupby('Month')[['Total']].sum().plot(kind = 'bar')
```

```python
print('El mes mas donde mas cierculas es: ')
df.groupby('Month')[['Total']].sum().sort_values(ascending = False, by = 'Total')[:1]
```

```python
df = pd.read_csv('311-service-requests.csv', low_memory = False)
df.head(5)
```

```python
# Cual es el tipo de queja mas comun?
df['Complaint Type'].value_counts()[0:1]
```

```python
# Grafico de las 10 mayores quejas
df['Complaint Type'].value_counts()[:10].plot(kind = 'bar')
```

```python
# Quejas de Brooklyn
quejas_brooklyn = df[(df['Borough'] == 'BROOKLYN') & (df['Complaint Type'] == 'Noise - Street/Sidewalk')]
```

```python
print(f'Hay {quejas_brooklyn.shape[0]} quejas en Brooklyn.')
```

```python
quejas_brooklyn['Borough'].value_counts()
```

```python
# Motivo principal de las quejas de Brooklyng
quejas_brooklyn.Descriptor.value_counts()
```

```python
# Zonas donde se dan mas quejas - Top 3
quejas_brooklyn['Incident Zip'].value_counts()[:3]
```
